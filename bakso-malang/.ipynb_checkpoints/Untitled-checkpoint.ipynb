Importing Review Data

```python
# The places
places = ["pahlawan-trip","pak-man","trowulan","damas","gun","cak-man","samut","presiden"]

# The reviews
reviews = []

for i, c in enumerate(places):
    with open("reviews/" + c + ".txt", "rb") as file:
        reviews.append(file.read())
```

Cleaning Review Data

```python
import re
import string

clean_phase_1 = []

def clean_text_phase1(text):
    text = text.decode('utf-8')
    
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('\r', ' ', text)
    text = re.sub('\n', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

for review in reviews:
    clean_phase_1.append(clean_text_phase1(review))
    
reviews = clean_phase_1

reviews
```
